```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from scipy import stats
```

```python
# Load the data
data = pd.read_csv("dataframe_.csv")
```

```python
# Handle missing values
imputer = SimpleImputer()
data = imputer.fit_transform(data)

# Outlier detection and treatment
z = np.abs(stats.zscore(data))
threshold = 3
data = data[(z < threshold).all(axis=1)]
# Split the data into X and y
X = data.drop("output", axis=1)
y = data["output"]
# Feature engineering
ct = ColumnTransformer(
    [("one_hot_encoder", OneHotEncoder(), [0, 1, 2]),],
    remainder="passthrough",
)

# Hyperparameter tuning
pipeline = Pipeline([("transformer", ct), ("scaler", StandardScaler()), ("regressor", RandomForestRegressor())])

param_grid = {
    "regressor__n_estimators": [100, 500, 1000],
    "regressor__max_depth": [3, 5, 7],
    "regressor__min_samples_split": [2, 3, 4],
}

grid = GridSearchCV(pipeline, param_grid, cv=5)
grid.fit(X, y)

# Evaluate the model
print("Best parameters: ", grid.best_params_)
print("Best score: ", grid.best_score_)
```
